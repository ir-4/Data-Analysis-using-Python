# Data Analysis & Visualization

**Product Category Analysis**

```python
#sales
crev = data.groupby(['Category'])['Sales'].sum().reset_index()
mycolors=['#4caf50','#2e6930','#1e4620']
myexplode=(0,0,0.1)
plt.subplot(1, 2, 1)
plt.pie(crev['Sales'], explode=myexplode, colors = mycolors, startangle = 90, autopct='%1.1f%%')
plt.title('Category-wise sales distribution') 

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=1.5,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

#orders
cord = data.groupby(['Category'])['Order_ID'].count().reset_index()
myexplode=(0,0.1,0)
plt.subplot(1, 2, 2)
plt.pie(cord['Order_ID'], colors = mycolors, explode=myexplode, startangle = 90, autopct='%1.1f%%')
plt.legend(cord['Category'], bbox_to_anchor=(1,1), loc='upper left', borderaxespad=0)
plt.title('Order count based on Category')
plt.show() 
```

Insights:<br>
    - Though Office Supplies accounts for the most orders, they are the least revenue generating category among the three.<br>
    - Though Technology accounts for the least orders, they are the most revenue generating category among the three.<br>
    - This can be due to the Price of the products in each category

**Region-wise distribution**

```python
#sales
rrev = data.groupby(['Region'])['Sales'].sum().reset_index()
mycolors=['#45b6fe','#3792cb','#296d98','#1c4966']
myexplode=(0,0,0,0.1)
plt.subplot(1, 2, 1)
plt.pie(rrev['Sales'], explode = myexplode, colors = mycolors, startangle = 90, autopct='%1.1f%%')
plt.title('Region-wise sales distribution')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=1.5,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

#orders
rord = data.groupby(['Region'])['Order_ID'].count().reset_index()
myexplode=(0,0,0,0.1)
plt.subplot(1, 2, 2)
plt.pie(rord['Order_ID'], colors = mycolors, explode=myexplode, startangle = 90, autopct='%1.1f%%')
plt.legend(rord['Region'], bbox_to_anchor=(1,1), loc='upper left', borderaxespad=0)
plt.title('Order count based on Region')
plt.show() 
```

Insights:<br>
    - West region performs well in orders as well as sales.<br>
    - South is the poor performer

**Customer Segments Analysis**

```python
srev = data.groupby(['Segment'])['Sales'].sum().reset_index()
mycolors=['#fff192','#ffea61','#ffdd3c']
myexplode=(0.1,0,0)
plt.subplot(1, 2, 1)
plt.pie(srev['Sales'], colors = mycolors, explode=myexplode, startangle = 90, autopct='%1.1f%%')
plt.title('Customer Segment-wise sales distribution')
#plt.show() 

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=1.5,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

#orders
sord = data.groupby(['Segment'])['Order_ID'].count().reset_index()
myexplode=(0,0,0)
plt.subplot(1, 2, 2)
myexplode=(0.1,0,0)
plt.pie(sord['Order_ID'], colors = mycolors, explode=myexplode, startangle = 90, autopct='%1.1f%%')
plt.legend(sord['Segment'], bbox_to_anchor=(1,1), loc='upper left', borderaxespad=0)
plt.title('Order count based on Segment')
plt.show() 
```

Insigths:<br>
    - Consumers share almost half of the sales as well as orders orders

**Top performing sub-categories**

```python
sc_rev = data.groupby(['Sub_Category'])['Sales'].sum().nlargest(10)
plt.figure(figsize=(25,5))
plt.bar(sc_rev.index,sc_rev.values,color='#a1045a')
plt.show()
```

**Trend over year**

```python
data['Order_Year'] = data['Order_Date'].dt.year
```

```python
data.head()
```

```python
#sales
y_rev = data.groupby(['Order_Year'])['Sales'].sum().reset_index()

default_x_ticks = range(len(y_rev['Order_Year']))
plt.subplot(1, 2, 1)
plt.plot(default_x_ticks, y_rev['Sales'], marker='o')
plt.xticks(default_x_ticks, y_rev['Order_Year'])
plt.title('Sales trend over year')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=1.5,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

#orders
y_ord = data.groupby(['Order_Year'])['Order_ID'].count().reset_index()
default_x_ticks = range(len(y_ord['Order_Year']))
plt.subplot(1, 2, 2)
plt.plot(default_x_ticks, y_ord['Order_ID'], marker='o')
plt.xticks(default_x_ticks, y_rev['Order_Year'])
plt.title('Order trend over year')
plt.show()  
```
